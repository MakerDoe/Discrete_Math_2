```python
import networkx as nx
import matplotlib.pyplot as plt

draw_params = {
    "with_labels": True,
    "edge_color": "gray",
    "width": 3,
    "linewidths": 3,
    "edgecolors": "gray",
    "node_color": "white",
    "node_size": 1000,
    "font_color": "black",
    "font_family": "monospace",
    "font_weight": "bold",
}
```

# Graph
A graph $G$ consists of two finite sets: a nonempty set $V(G)$ of vertices and a set $E(G)$ of edges, where each edge is associated with a set consisting of either one or two vertices called its endpoints. The correspondence from edges to endpoints is called the edge-endpoint function.

```python
G = nx.Graph()
G.add_edges_from([(1, 1), (1, 2), (2, 3), (2, 3), (3, 4), (4, 5), (5, 1), (5, 6)])

nx.draw(G, **draw_params)
plt.show()
```

# Directed Graph

---

A directed graph, or digraph, consists of two finite sets: a nonempty set $V(G)$ of vertices and a set $D(G)$ of directed edges, where each is associated with an ordered pair of vertices called its endpoints. If edge $e$ is associated with the pair $(v, w)$ of vertices, then e is said to be the $(directed)$ edge from $v$ to $w$.


```python
G_directed = nx.DiGraph()
G_directed.add_edges_from([(1, 2), (2, 3), (3, 1), (3, 4), (4, 5), (5, 1)])

nx.draw(G_directed, **draw_params, arrowsize=40)
plt.show()
```

# Simple Graph

---

A simple graph is a type of graph in graph theory characterized by two specific properties:

1. **No Loops:** A simple graph does not have any edges that connect a vertex to itself. This means there are no edges that start and end at the same vertex.

2. **No Multiple Edges:** A simple graph does not have multiple edges between any pair of vertices. There is at most one edge connecting any two distinct vertices.

```python
G_simple = nx.Graph()
G_simple.add_edges_from([(1, 2), (2, 3), (3, 1)])

nx.draw(G_simple, **draw_params)
plt.show()
```

# Complete Graph

---

A complete graph is a specific type of graph in graph theory where every pair of distinct vertices is connected by a unique edge. In other words, it is a simple graph where every vertex is directly connected to every other vertex by exactly one edge.


```python
G_complete = nx.complete_graph(8)

nx.draw(G_complete, **draw_params)
plt.show()
```

# Bipartite Graph

---

A bipartite graph is a graph whose vertices can be divided into two distinct sets such that no edges exist within the same set. This structure makes bipartite graphs particularly useful for modeling relationships between two different types of entities and for solving various combinatorial and optimization problems.

```python
G_bipartite = nx.Graph()
G_bipartite.add_nodes_from([1, 2, 3, 4], bipartite=0)
G_bipartite.add_nodes_from(["A", "B", "C"], bipartite=1)
G_bipartite.add_edges_from(
    [(1, "A"), (1, "B"), (2, "A"), (2, "B"), (2, "C"), (3, "A"), (4, "C")]
)

nx.draw(G_bipartite, nx.bipartite_layout(G_bipartite, [1, 2, 3, 4]), **draw_params)
plt.show()
```

# Subgraph

---

A subgraph is a graph formed from a subset of the vertices and edges of a larger graph. Depending on how these subsets are chosen, subgraphs can be induced, spanning, or edge-induced, each useful for different analyses and applications in graph theory and network studies.

```python
_, axes = plt.subplots(1, 2, figsize=(12, 6))

G_main = nx.complete_graph(6)
G_sub = nx.subgraph(G_main, [0, 1, 2])

nx.draw(G_main, ax=axes[0], **draw_params)
axes[0].set_title("Original Graph")

nx.draw(G_sub, ax=axes[1], **draw_params)
axes[1].set_title("Subgraph")

plt.tight_layout()
plt.show()
```

# The Handshake Theorem
---
The Handshake Theorem is a simple yet powerful tool in graph theory, highlighting the relationship between vertex degrees and edges in an undirected graph. It ensures the consistency of graph structures and aids in the analysis and understanding of various graph properties.

- **Degree of a Vertex:** The degree of a vertex is the number of edges incident to (connected to) that vertex.
- **Sum of Degrees:** When you sum the degrees of all vertices, each edge is counted twice (once for each endpoint of the edge). Thus, the total sum of the degrees is twice the number of edges.

# Travelling in Graph
---
"Travelling in a graph" typically refers to various concepts related to moving through a graph from one vertex to another. This can involve paths, walks, trails, circuits, or cycles, depending on the context and specific constraints. Here are some key concepts related to traveling in a graph:

1. **Walk:** A walk is a sequence of vertices and edges, where vertices and edges can be repeated.
2. **Trail:** A trail is a walk in which no edge is repeated, though vertices can be repeated.
3. **Path:** A path is a sequence of edges that connects a sequence of distinct vertices.
4. **Closed Walk:** A closed walk is a walk from $v$ to $w$ that starts and ends in the same vertex.
5. **Circuit:** A circuit is a trail that starts and ends at the same vertex.
6. **Cycle:** A cycle is a path (no vertices repeated) that starts and ends at the same vertex.

# Connectedness of a Graph
---
Connectedness is a fundamental concept in graph theory that describes the extent to which the vertices of a graph are reachable from one another. Understanding whether a graph is connected, and identifying its connected components, are crucial steps in analyzing the structure and properties of the graph.

# Connected Components
---
A graph $H$ is connected component of a graph $G$ $\Longleftrightarrow$:

- $H$ is a subgraph of $G$.
- $H$ is connected.
- No connected subgraphs of $G$ has $H$ as a subgraphs and contains vertices or edges that are not in $H$.

# Euler's Circuit

---

An Eulerian circuit is a closed walk in an undirected graph that traverses each edge exactly once and returns to the starting vertex. Understanding the properties and existence of Eulerian circuits is essential in various fields, including network design, circuitry, and optimization problems like the Traveling Salesman Problem.

An algorithm called Fleury's algorithm can be used to find an Eulerian circuit in a graph. This algorithm is based on the idea of traversing the graph while avoiding bridges (edges that, when removed, disconnect the graph) until all edges have been visited.

```python
G_euler_circuit = nx.cycle_graph(5)
nx.draw(G_euler_circuit, nx.circular_layout(G_euler_circuit), **draw_params)
plt.show()
```

# Euler's Trail

---

An Eulerian trail is a non-closed walk in a graph that traverses each edge exactly once but may not necessarily return to the starting vertex. Understanding the properties and existence of Eulerian trails is important in various fields, including network design, circuitry, and optimization problems.

Finding an Eulerian trail in a graph can be more challenging than finding an Eulerian circuit. However, it can be done using algorithms such as Fleury's algorithm or Hierholzer's algorithm, which are designed to efficiently find Eulerian trails or circuits in graphs.

```python
G_euler_trail = nx.path_graph(6)
nx.draw(G_euler_trail, nx.circular_layout(G_euler_trail), **draw_params)
plt.show()
```

# Hamiltonian Circuit

---

A Hamiltonian circuit is a closed walk in a graph that visits every vertex exactly once, except for the starting vertex, which is visited again at the end of the circuit. Understanding the properties and existence of Hamiltonian circuits is important in various fields, including optimization problems like the Travelling Salesman Problem and network design.

Unlike Eulerian circuits, finding Hamiltonian circuits in graphs is a much harder problem and there is no known polynomial-time algorithm that can solve it for all graphs. Many special cases and heuristics have been developed to address specific instances of the problem, but finding Hamiltonian circuits in general graphs remains a challenging task.


```python
G_hamiltonian = nx.cycle_graph(6)
nx.draw(G_hamiltonian, nx.circular_layout(G_hamiltonian), **draw_params)
plt.show()
```
